```python
import pandas as pd
import numpy as np

df = pd.read_csv("student_health_lifestyle_dataset.csv")
df.head()
```

# 1) Missing Values Checking ::

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# Sleep_Hours (Numerical) → we take (Mean) 

mean_sleep = df['Sleep_Hours'].mean()
df['Sleep_Hours'].fillna(mean_sleep, inplace=True)

print("\nAfter filling Sleep_Hours with Mean:\n", df[['Student_ID','Sleep_Hours']])

```

```python
# Calories_Intake (Numerical) → we take (Median) 

median_calories = df['Calories_Intake'].median()
df['Calories_Intake'].fillna(median_calories, inplace=True)

print("\nAfter filling Calories_Intake with Median:\n", df[['Student_ID','Calories_Intake']])
```

```python
# Exercise_Hours (Numerical) → we take (Mean) 

mode_exercise = df['Exercise_Hours'].mean()
df['Exercise_Hours'].fillna(mode_exercise, inplace=True)

print("\nAfter filling Exercise_Hours with Mean:\n", df[['Student_ID','Exercise_Hours']])

```

```python
# BMI (Numerical) → we take (Interpolation)

df['BMI'] = df['BMI'].interpolate()

print("\nAfter filling BMI using Interpolation:\n", df[['Student_ID','BMI']])

```

# 2) Check for Duplicates

```python
print("Number of duplicate rows:", df.duplicated().sum())
```

# 3) Obtain unique values of a column 

```python
# Obtain unique values for all columns
df['Gender'].unique()
```

# 4) Make textual data uniform in case (Gender → lowercase)

```python
# 4) Make textual data uniform in case (Gender → lowercase)

df['Gender'] = df['Gender'].str.lower()

print("\nAfter making Gender lowercase:\n", df[['Student_ID','Gender']])
```

# 5) Identify and remove the outliers

```python
numeric_cols = ['Age','Sleep_Hours','Calories_Intake','Stress_Level',
                'Exercise_Hours','BMI','Screen_Time','Exam_Score']

# Looping in numeric columns and removing outliers
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]

print("\nDataset after removing outliers:\n", df)
```

# 6) Perform Data Normalization (Min-Max Scaling)

```python
from sklearn.preprocessing import MinMaxScaler

# Numeric columns to normalize
numeric_cols = ['Age','Sleep_Hours','Calories_Intake','Stress_Level',
                'Exercise_Hours','BMI','Screen_Time','Exam_Score']

scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("\nDataset after Min-Max Normalization:\n", df)
```

# 7) Perform Data Standardization (Z-score Scaling)

```python
from sklearn.preprocessing import StandardScaler

# Numeric columns to standardize
numeric_cols = ['Age','Sleep_Hours','Calories_Intake','Stress_Level',
                'Exercise_Hours','BMI','Screen_Time','Exam_Score']

scaler = StandardScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("\nDataset after Standardization (Z-score Scaling):\n", df)
```
