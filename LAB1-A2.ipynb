## COVID DATASET:::

```python
import pandas as pd
import numpy as np

df = pd.read_csv("covid.csv")
df.head()
```

# 1) Missing Values Checking ::

```python
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# Lat (Numerical) → we take (Mean)

mean_lat = df['Lat'].mean()
df['Lat'].fillna(mean_lat, inplace=True)

print("\nAfter filling Lat with Mean:\n", df[['Country_Region','Lat']])
```

```python
# Long_ (Numerical) → we take (Mean)

mean_long = df['Long_'].mean()
df['Long_'].fillna(mean_long, inplace=True)

print("\nAfter filling Long_ with Mean:\n", df[['Country_Region','Long_']])
```

```python
# Confirmed (Numerical) → we take (Median)

median_confirmed = df['Confirmed'].median()
df['Confirmed'].fillna(median_confirmed, inplace=True)

print("\nAfter filling Confirmed with Median:\n", df[['Country_Region','Confirmed']])
```

```python
# Deaths (Numerical) → we take (Median)

median_deaths = df['Deaths'].median()
df['Deaths'].fillna(median_deaths, inplace=True)

print("\nAfter filling Deaths with Median:\n", df[['Country_Region','Deaths']])
```

```python
# Active (Numerical) → we take (Mean)

mode_active = df['Active'].mean()
df['Active'].fillna(mode_active, inplace=True)

print("\nAfter filling Active with Mean:\n", df[['Country_Region','Active']])
```

```python
# Incident_Rate (Numerical) → Interpolation

df['Incident_Rate'] = df['Incident_Rate'].interpolate()

print("\nAfter filling Incident_Rate using Interpolation:\n", df[['Country_Region','Incident_Rate']])
```

```python
# Case_Fatality_Ratio (Numerical) → Interpolation

df['Case_Fatality_Ratio'] = df['Case_Fatality_Ratio'].interpolate()

print("\nAfter filling Case_Fatality_Ratio using Interpolation:\n", df[['Country_Region','Case_Fatality_Ratio']])
```

# 2 ) Check Duplicate Values ::

```python
print("Number of duplicate rows:", df.duplicated().sum())
```

```python
# Removing duplicate rows

df.drop_duplicates(inplace=True)
print("\nDataset after removing duplicates:\n", df)
```

```python
print("Number of duplicate rows after:", df.duplicated().sum())
```

```python
df['Country_Region'].unique()
```

# 4) Make textual data uniform in case 

```python
# Here COuntry region

df['Country_Region'] = df['Country_Region'].str.lower()

print("\nAfter making Country_Region lowercase:\n", df[['Country_Region']])
```

# 5) Identify and remove the outliers

```python
# Numeric columns in Covid dataset
numeric_cols = ['Lat','Long_','Confirmed','Deaths','Recovered','Active',
                'Incident_Rate','Case_Fatality_Ratio']

# Create an empty DataFrame to store outliers
outliers_df = pd.DataFrame()

# Loop over numeric columns
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    # Select outliers for this column
    col_outliers = df[(df[col] < lower) | (df[col] > upper)]
    
    # Append to the outliers DataFrame
    outliers_df = pd.concat([outliers_df, col_outliers]).drop_duplicates()
    
    # Remove outliers from the original dataset
    df = df[(df[col] >= lower) & (df[col] <= upper)]

print("Dataset after removing outliers:\n", df)
print("\nOutliers that were removed:\n", outliers_df)
```

# 6) Perform Data Normalization (Min-Max Scaling)

```python
# 6) Perform Data Normalization (Min-Max Scaling) for numeric columns

from sklearn.preprocessing import MinMaxScaler

# Numeric columns
numeric_cols = ['Lat','Long_','Confirmed','Deaths','Recovered','Active',
                'Incident_Rate','Case_Fatality_Ratio']

scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("\nDataset after Min-Max Normalization:\n", df)
```

# Perform Data Standardization (Z-score Scaling)

```python
from sklearn.preprocessing import StandardScaler

# Numeric columns
numeric_cols = ['Lat','Long_','Confirmed','Deaths','Recovered','Active',
                'Incident_Rate','Case_Fatality_Ratio']

# Initialize StandardScaler
scaler = StandardScaler()

# Apply Z-score scaling
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

print("\nDataset after Standardization (Z-score Scaling):\n", df)
```
