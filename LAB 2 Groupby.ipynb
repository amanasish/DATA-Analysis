# Groupby 

```python
import pandas as pd
```

```python
df = pd.read_csv("covid.csv")
```

```python
df.head()
```

```python
df.Lat.isna().sum()
```

```python
df.Country_Region.unique()
```

```python
country_dfs = dict(tuple(df.groupby('Country_Region')))
```

```python
country_dfs["US"]
```

```python
num_cols = df.select_dtypes(include='number').columns
```

```python
# Mean
for country, cdf in country_dfs.items():
    cdf[num_cols] = cdf[num_cols].fillna(cdf[num_cols].mean())
```

```python
df_final = pd.concat(country_dfs.values(),ignore_index=True)
```

```python
df_final
```

```python
df.isna().sum()
```

```python
df_final.isna().sum()
```

```python
# Median
for country, cdf in country_dfs.items():
    cdf[num_cols] = cdf[num_cols].fillna(cdf[num_cols].median())
    

df_final = pd.concat(country_dfs.values())
df_final
```

```python
df_final.isna().sum()
```

```python
# Interpolation
for country, cdf in country_dfs.items():
    cdf[num_cols] = cdf[num_cols].interpolate(method='linear')

df_final = pd.concat(country_dfs.values())
df_final
```

```python
df_final.isna().sum()
```

```python
# 1. Find the top 3 genres by total earning

genre_earnings = df2.groupby('Genre')['Gross'].sum()
genre_earnings = genre_earnings.sort_values(ascending=False)
top3_genres = genre_earnings.head(3)
```
